```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot the eigenvalues of 2D model problem when omega = 2/3, 4/5, 1
n = 16
omega = [2./3., 4./5., 1.]
k = np.linspace(0,1,n+1)[1:-1]
l = np.linspace(0,1,n+1)[1:-1]
k,l = np.meshgrid(k,l)
eigen_R0 = np.abs(1-omega[0]*(np.power(np.sin(np.pi*k/2.),2)+np.power(np.sin(np.pi*l/2.),2)))
eigen_R1 = np.abs(1-omega[1]*(np.power(np.sin(np.pi*k/2.),2)+np.power(np.sin(np.pi*l/2.),2)))
eigen_R2 = np.abs(1-omega[2]*(np.power(np.sin(np.pi*k/2.),2)+np.power(np.sin(np.pi*l/2.),2)))
```

```python
# k,l~1 is high frequency; k,l~0 is low frequency

fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
surf0 = ax.plot_surface(k,l,eigen_R0,label='omega = 2/3')
surf0._facecolors2d=surf0._facecolor3d
surf0._edgecolors2d=surf0._edgecolor3d
surf1 = ax.plot_surface(k,l,eigen_R1,label='omega = 4/5')
surf1._facecolors2d=surf1._facecolor3d
surf1._edgecolors2d=surf1._edgecolor3d
surf2 = ax.plot_surface(k,l,eigen_R2,label='omega = 1')
surf2._facecolors2d=surf2._facecolor3d
surf2._edgecolors2d=surf2._edgecolor3d
ax.legend(handles=[surf0,surf1,surf2])
```

```python
# 2D plots, k components are for middle frequency, change l from middle to high
ll = np.linspace(0,1,n+1)[1:-1]
kk = ll[7]
eigen_2d0 = np.abs(1-omega[0]*(np.power(np.sin(np.pi*kk/2.),2)+np.power(np.sin(np.pi*ll/2.),2)))
eigen_2d1 = np.abs(1-omega[1]*(np.power(np.sin(np.pi*kk/2.),2)+np.power(np.sin(np.pi*ll/2.),2)))
eigen_2d2 = np.abs(1-omega[2]*(np.power(np.sin(np.pi*kk/2.),2)+np.power(np.sin(np.pi*ll/2.),2)))
plt.plot(ll,eigen_2d0,label="omega=2/3")
plt.plot(ll,eigen_2d1,label="omega=4/5")
plt.plot(ll,eigen_2d2,label="omega=1")
plt.legend()
plt.xlabel('l')
plt.ylabel('$\lambda$(R)')
```

```python
# 2D plots, max eigenvalues when changing k components for fix l
l = np.linspace(0,1,n+1)[1:-1]
eigen_R0l = eigen_R0.max(axis=1)
eigen_R1l = eigen_R1.max(axis=1)
eigen_R2l = eigen_R2.max(axis=1)

plt.plot(l[7:],eigen_R0l[7:],label='omega=2/3')
plt.plot(l[7:],eigen_R1l[7:],label='omega=4/5')
plt.plot(l[7:],eigen_R2l[7:],label='omega=1')
plt.title("Maximum eigenvalues for high frequency modes")
plt.legend()
```
