Generate dataset for 2D Poisson periodic problems

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import h5py
```

```python
def rand_rhs(n):
    x = np.linspace(-1,1,n)
    y = np.linspace(-1,1,n)
    xx, yy = np.meshgrid(x, y, indexing='ij')

    k = 20.*random.random() - 10. # float number in range of [-10, 10)
    l = 1 #random.randint(-6,6).__float__() # int number in range of [-6,6]
    m = 1 #random.randint(-6,6).__float__() # int number in range of [-6,6]

    rhs = k*np.sin(l*np.pi*(xx+0.5))*np.cos(m*np.pi*(yy)).reshape((1,1,n,n))

    return rhs
```

```python
# Create dataset
n = 2**5

dataset = {}
n_data = 1000
f_data = np.zeros((n_data, n+1, n+1))
for i in range(n_data):
    f = rand_rhs(n+1)
    f_data[i,:,:] = f.reshape((n+1, n+1))
    print("Data #-"+str(i)+" generated")
```

```python
plt.imshow(f_data[0])
```

```python
dataset['rhs'] = f_data

with h5py.File("poisson2d_33x33_pbc.h5", "w") as hf:
    for key, item in dataset.items():
        hf[key] = item
```
